```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import os
import pickle
import pickle_readers
plt.rcParams['figure.figsize'] = [4, 4]
plt.rcParams['figure.dpi'] = 250
plt.rcParams['font.size'] = 4
np.set_printoptions(linewidth=200,precision=10,suppress=True)
```

```python
cwd = os.getcwd()
save_folder = os.path.join(os.path.join(cwd,"Prior_Runs"),"functional_arc_2")
pickle_path = os.path.join(save_folder,"Positions_History.pickle")
reader = pickle_readers.positions_reader(pickle_path)
result = reader.read()
with open(os.path.join(save_folder,"pure_positions.pickle"),'wb') as handle:
    pickle.dump(result, handle)
```

```python
fig,ax = plt.subplots()
for each in result.values():
    ax.scatter(each["POS_X"],each["POS_Y"],s=0.1,lw=0)
ax.set_xlim([-500,500])
ax.set_ylim([-500,500])
ax.set_title("No communication functional arc")
```
